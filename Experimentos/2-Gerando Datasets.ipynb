```python
import pandas as pd
import arff # LIAC-ARFF
from utils import fix_arff
from os import listdir, makedirs
```

```python
datasets = listdir('classification/')
```

```python
import os
```

```python
for dataset in datasets:
    print('\n', dataset)
    
    for i in range(1, 11):
        print(i, end=' ')
        treino_file = open(f'classification/{dataset}/{dataset}-10-{i}tra.dat', 'r').read().split('\n')
        fix_arff(treino_file)
        treino = arff.load('\n'.join(treino_file))


        test_file = open(f'classification/{dataset}/{dataset}-10-{i}tst.dat', 'r').read().split('\n')
        fix_arff(test_file)
        teste = arff.load('\n'.join(test_file))

        treino = pd.DataFrame(treino['data'],columns=[x[0] for x in treino['attributes']])
        teste = pd.DataFrame(teste['data'],columns=[x[0] for x in teste['attributes']])
        
        
        treino.to_excel(f'datasets/{dataset}/{dataset}-{i}-train.xlsx', index=False)
        teste.to_excel(f'datasets/{dataset}/{dataset}-{i}-test.xlsx', index=False)
```
